```python
import pandas
```

```python
merged_dataframe = pandas.read_csv("merged_stocks_and_twitter_sentiment.csv")
```

```python
merged_dataframe
```

```python
dataframe = merged_dataframe[["Date",
                             "Adj Close",
                             "Volume",
                             "ts_polarity",
                             "twitter_volume"]]
```

```python
dataframe.set_index("Date", inplace = True)
```

```python
dataframe
```

```python
sentiments = []
```

```python
POSITIVE_SENTIMENT_THRESHOLD = 0.05
```

```python
NEGATIVE_SENTIMENT_THRESHOLD = -0.05
```

```python
for score in dataframe["ts_polarity"]:
    if score >= POSITIVE_SENTIMENT_THRESHOLD:
        sentiments.append("Positive")
    elif score <= NEGATIVE_SENTIMENT_THRESHOLD:
        sentiments.append("Negative")
    else :
        sentiments.append("Neutral")
```

```python
dataframe["Sentiment"] = sentiments
```

```python
dataframe
```

```python
dataframe["Sentiment"].value_counts()
```

```python
dataframe["Price Difference"] = dataframe["Adj Close"].diff()
dataframe.dropna(inplace =True)
dataframe
```

```python
import numpy
```

```python
RISE = 1
FALL = 0
dataframe ["Stock Trend"] = numpy.where(
    dataframe["Price Difference"]> 0, RISE, FALL)
dataframe
```

```python
new_dataframe = dataframe [["Adj Close",
                           "Volume",
                           "twitter_volume",
                          "Sentiment",
                          "Stock Trend"]]
```

```python
new_dataframe = pandas.get_dummies(new_dataframe, columns = ["Sentiment"])
```

```python
new_dataframe
```

```python
x = new_dataframe.copy()
x.drop("Stock Trend", axis=1 , inplace=True)
x
```

```python
y = new_dataframe["Stock Trend"].values.reshape(-1,1)
y[:10]
```

```python
SPLIT = int (0.8*len(x))
x_train = x[:SPLIT]

x_test = x[SPLIT:]

y_train = y[:SPLIT]
y_test = y[SPLIT:]
```

```python
y_test
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_scaler = scaler.fit(x_train)
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled = x_scaler.transform(x_test)

x_train_scaled
```

```python
from sklearn.ensemble import RandomForestClassifier
NUMBER_OF_ESTIMATORS = 500
RANDOM_STATE = 80

classifier = RandomForestClassifier(n_estimators= NUMBER_OF_ESTIMATORS,
                                   random_state=RANDOM_STATE)

```

```python
x_test_scaled
```

```python
y_train
```

```python
y_train.ravel()
```

```python
classifier = classifier.fit(x_train_scaled,y_train.ravel())

```

```python
classifier.warm_start
```

```python
predictions = classifier.predict(x_test_scaled)
```

```python
predictions
```

```python
pandas.DataFrame({
    "Predictions": predictions,
    "Actual": y_test.ravel()
})
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix_results = confusion_matrix(y_test, predictions)
confusion_matrix_dataframe = pandas.DataFrame(confusion_matrix_results,
                index = ["Actual 0", "Actual 1"],
                columns= ["Predicted 0",  "Predicted 1"])

display(confusion_matrix_dataframe)
```

```python
from sklearn.ensemble import GradientBoostingClassifier


NUMBER_OF_ESTIMATORS = 30
MAX_NUMBER_OF_FEATURES = 2
MAX_DEPTH=3
RANDOOM_STATE_CONSTANT =0

learning_rates = [0.05,0.1,0.25,0.5,0.75,1,1.5,2]

for learning_rate in learning_rates:
    classifier = GradientBoostingClassifier(n_estimators= NUMBER_OF_ESTIMATORS,
                                           learning_rate= learning_rate,
                                           max_features= MAX_NUMBER_OF_FEATURES,
                                           max_depth=MAX_DEPTH,
                                           random_state=RANDOOM_STATE_CONSTANT)
    
    classifier.fit(x_train_scaled, y_train.ravel())
    
    training_score = classifier.score(x_train_scaled,y_train.ravel())
    testing_score = classifier.score(x_test_scaled,y_test.ravel())
    
    print("Learning rate: ", learning_rate)
    print("Training Accuracy: ", training_score)
    print("Testing Accuracy: ", testing_score)
    print()
    
    
    
 
```

```python
LEARNING_RATE = 1

MAX_NUMBER_OF_FEATURES = 5


classifier = GradientBoostingClassifier(n_estimators= NUMBER_OF_ESTIMATORS,
                                      learning_rate=LEARNING_RATE,
                                      max_features=MAX_NUMBER_OF_FEATURES,
                                      max_depth=MAX_DEPTH,
                                      random_state=RANDOOM_STATE_CONSTANT
                                      )

classifier.fit(x_train_scaled, y_train.ravel())

training_score = classifier.score(x_train_scaled,y_train.ravel())
testing_score = classifier.score(x_test_scaled,y_test.ravel())

print("Learning rate: ", LEARNING_RATE)
print("Training Accuracyy: ", training_score)
print("Testing Accuracy: ", testing_score)
print()
```

```python
predictions = classifier.predict(x_test_scaled)
pandas.DataFrame({
    "Prediction": predictions,
    "Actual Stock Trend": y_test.ravel()
    
})
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Prediction Model Accuracy: ", accuracy)
```

```python
print(classification_report(y_test,predictions))
```
